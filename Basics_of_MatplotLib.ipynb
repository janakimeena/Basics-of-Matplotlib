```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
```

```python
plt.plot(x,y)
plt.title('Our First Graph!')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

### Add X and Y Label

```python
plt.plot(x,y)
plt.title('Our First Graph!',fontdict={'fontname':'Sans','fontsize':20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

### Add ticks, Change fontname and size of title

```python
plt.plot(x,y)
plt.title('Our First Graph!',fontdict={'fontname':'Sans','fontsize':20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.show()
```

### Add legent to the chart, add label to graph

```python
plt.plot(x,y,label = '2x')
plt.title('Our First Graph!',fontdict={'fontname':'Sans','fontsize':20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

### Change color of line

```python
plt.plot(x,y,label = '2x',color = 'g') # color can be given as #ffffff .... or single letter
plt.title('Our First Graph!',fontdict={'fontname':'Sans','fontsize':20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

### Specify line width

```python
plt.plot(x,y,label = '2x',color = 'g',linewidth=3) # color can be given as #ffffff .... or single letter
plt.title('Our First Graph!',fontdict={'fontname':'Sans','fontsize':20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

### Add markers and cutomize look of marker

```python
plt.plot(x,y,label = '2x',color = 'g',linewidth=3, marker='.',markersize=10, markeredgecolor='blue') 
# color can be given as #ffffff .... or single letter
plt.title('Our First Graph!',fontdict={'fontname':'Sans','fontsize':20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

### Change line style

```python
plt.plot(x,y,label = '2x',color = 'g',linewidth=3,  marker='.',markersize=10, markeredgecolor='blue',linestyle='--') 
# color can be given as #ffffff .... or single letter
plt.title('Our First Graph!',fontdict={'fontname':'Sans','fontsize':20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

### Shorthand for color, marker, line

```python
# plt.plot(x,y,label = '2x',color = 'g',linewidth=3,  marker='.',markersize=10, markeredgecolor='blue',linestyle='--') 
# use shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y,'b^--',label = '2x')
# color can be given as #ffffff .... or single letter
plt.title('Our First Graph!',fontdict={'fontname':'Sans','fontsize':20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

### Add a Second Line

```python
# plt.plot(x,y,label = '2x',color = 'g',linewidth=3,  marker='.',markersize=10, markeredgecolor='blue',linestyle='--') 
# use shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y,'b^--',label = '2x')
# color can be given as #ffffff .... or single letter
## Line Number Two
x2 = np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,c='r',label='x^2')
plt.title('Our First Graph!',fontdict={'fontname':'Sans','fontsize':20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

### Change line syle for part of a code

```python
# plt.plot(x,y,label = '2x',color = 'g',linewidth=3,  marker='.',markersize=10, markeredgecolor='blue',linestyle='--') 
# use shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y,'b^--',label = '2x')
# color can be given as #ffffff .... or single letter
## Line Number Two
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,c='r',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'r--')
plt.title('Our First Graph!',fontdict={'fontname':'Sans','fontsize':20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

### Resize Figure

```python
# resize a graph
plt.figure(figsize=(5,3),dpi=120)
plt.plot(x,y,'b^--',label = '2x')
# color can be given as #ffffff .... or single letter
## Line Number Two
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,c='r',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'r--')
plt.title('Our First Graph!',fontdict={'fontname':'Sans','fontsize':20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

### Save a graph

```python
# resize a graph
plt.figure(figsize=(5,3),dpi=120)
plt.plot(x,y,'b^--',label = '2x')
# color can be given as #ffffff .... or single letter
## Line Number Two
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,c='r',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'r--')
plt.title('Our First Graph!',fontdict={'fontname':'Sans','fontsize':20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.savefig('mygraph.png',dpi=300) # greater dpi will have good print
plt.show()
```

### Bar charts

```python
labels = ['A','B','C']
values = [1,2,3]
plt.bar(labels,values)
plt.figure(figsize=(6,4),dpi=100)
plt.show()
```

# Add pattern to first bar

```python
labels = ['A','B','C']
values = [1,2,3]
bars = plt.bar(labels,values)
bars[0].set_hatch('/')
plt.figure(figsize=(6,4),dpi=100)
plt.show()
```

### Add pattern to all bars

```python
labels = ['A','B','C']
values = [1,2,3]
bars = plt.bar(labels,values)
bars[0].set_hatch('/')
bars[1].set_hatch('o')
plt.figure(figsize=(6,4),dpi=100)
plt.show()
```

```python
labels = ['A','B','C']
values = [1,2,3]
bars = plt.bar(labels,values)
bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')
plt.figure(figsize=(6,4),dpi=100)
plt.show()
```

### Method 2 to add patterns

```python
labels = ['A','B','C']
values = [1,2,3]
bars = plt.bar(labels,values)
patterns = ['/','o','*']
for i,bar in enumerate(bars):
  bar.set_hatch(patterns[i])
plt.figure(figsize=(6,4),dpi=100)
plt.show()
```
